```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country =['India','Japan','Mexico','Colombia','Germany']
population =[1000,800,900,1000,300]
```

```python
cal=np.array(population)

prom = cal.mean()
dest = cal.std()
lim_sup = prom +3*dest
lim_inf = prom - 3*dest
```

```python
plt.bar(country,population)
plt.show()
```

```python
plt.barh(country,population)
plt.show()
```
