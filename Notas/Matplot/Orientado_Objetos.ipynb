# Graficos Orientados a objetos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
# Crea un objeto de tipo lienzo.
fig = plt.figure() 

# Crear un espacio para contener un grfico.
# fig.add([Posicin,Tamao Vertical del Grfico, Tamao Horizontal del Grfico, Altura])
# Posicin
axes = fig.add_axes([0.1,0.1,0.8,0.9]) 
axes2 =fig.add_axes([0.15,0.68,0.4,0.3])

#Grfico
axes.plot(x,y,'b')
axes2.plot(y,x,'r')

#Modificaciones color, ejes, labels etc..
axes2.set_facecolor('gray')
```
