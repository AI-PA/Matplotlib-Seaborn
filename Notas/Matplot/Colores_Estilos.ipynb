```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
# Diferentes estilos que maneja matplotlib
print(plt.style.available)
```

```python
plt.style.use('bmh')
```

## Matlab

```python
fig, ax = plt.subplots(figsize =(8,8))

ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'go:')
```

## Pyplot

```python
fig, ax = plt.subplots(figsize =(8,8))

ax.plot(x,x+1, color ='green',alpha=0.5, linewidth=5, linestyle='-')
ax.plot(x,x+2, color ='blue',linestyle='--')
ax.plot(x,x+3, color ='#66FF66',linestyle='dashed')
ax.plot(x,x+4, color ='#CCFFCC',linestyle=':')
```

```python
for x in plt.style.available:
    plt.style.use(x)
    fig, ax = plt.subplots(figsize=(2,2))
    ax.set_title(x, size=20)
```
