# Matplot

```python
#Importando Librerias
import matplotlib.pyplot as plt
import numpy as np 
```

```python
#Definiendo Variables
x = np.linspace(0,5,11)
y = x**2
```

## Empezando a manipular la librer√≠a

```python
# Grafico de relacion 2 variables
plt.plot(x,y,'bx')
plt.show()
```

```python
plt.plot(x,y,'rs-')
plt.show()
```

```python
plt.plot(x,y,'yD:')
plt.show()
```

```python
# Histograma
plt.hist(x)
plt.show()
```

```python
# Grafico de Pastel
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```
