```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
z = np.cos(x)
```

```python
fig, axe = plt.subplots(1,2,figsize = (5,5))
# AXE 0 
axe[0].plot(x,y,label ="$sin(x)$")
axe[0].set_title("Relación X - Y")
axe[0].set_xlabel('x')
axe[0].set_ylabel('y')
axe[0].legend(loc='lower left')
# AXE 1
axe[1].plot(y,x)
axe[1].set_title("Relación Y - X")
axe[1].set_xlabel('y')
axe[1].set_ylabel('x')
```
