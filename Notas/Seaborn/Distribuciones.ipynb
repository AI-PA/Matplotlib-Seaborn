```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips,x='tip', bins=15, cumulative=False, hue='sex', stat='frequency')
# sns.histplot(data=tips,x='tip', bins=Num.Columnas , cumulative=Acumulado, hue=leyenda , stat= Acumulado por )
plt.show()
```

```python
sns.histplot(data=tips,x='tip', bins=15, cumulative=False, hue='sex', stat='percent')
plt.show()
```

```python
sns.histplot(data=tips,x='tip', bins=15, cumulative=False, hue='sex', stat='frequency', multiple='stack')
# multiple = como se organizan los datos agrupados, uno al lado del otro, fill. 
plt.show()
```

```python
sns.histplot(data=tips,x='tip', bins=15, cumulative=False, hue='sex', stat='frequency', multiple='dodge')
plt.show()
```

```python
sns.histplot(data=tips,x='tip', bins=15, cumulative=False, hue='sex', stat='frequency', multiple='fill')
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex', cumulative=False)
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex', cumulative=True)
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex', cumulative=False, fill=True)
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex', cumulative=False, fill=True, bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex', stat='count')
plt.show()
```

```python
sns.displot(data=tips,x='tip',hue='sex', stat='count',kind='hist')
plt.show()
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='kde')
plt.show()
```
