```python
# Carga de librerías
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga del archivo
df = pd.read_csv("./Data/StudentsPerformance.csv")
```

```python
# Tabla dinámica 
table=df.pivot_table(values=['math score','reading score','writing score'],index='race/ethnicity',aggfunc='mean')
table.reset_index(inplace=True)
```

```python
table
```

## Matplot Lib

```python

plt.figure(figsize=(5,5))
plt.plot(table['race/ethnicity'],table['math score'], 'o-', label='Prom Obtenido')
plt.title('Promedio de Calificaciones de Matemáticas por Grupo')
plt.xlabel('Grupos')
plt.ylabel('Calificación Obtenida')
plt.legend(loc='lower right')
```

```python
fig, ax = plt.subplots()
plt.scatter(df['writing score'],df['math score'], marker='*', alpha=0.3)
plt.title('Relacion Resultado de Escritura y Resultado Matematicas')
plt.xlabel('writing score')
plt.ylabel('math score')
plt.show()

```

```python
# Variable a graficar
x = np.arange(100)

# Definimos nuestra contenedor/subplot
fig,ax=plt.subplots(2,2,figsize=(15,15))


#Gráfico superior izquierdo
ax[0,0].plot(x,np.sin(x), label = 'sin(x)')
ax[0,0].set_title("Sin(x)")
ax[0,0].set_xlabel('X')
ax[0,0].set_ylabel('Y')
ax[0,0].legend(loc='upper center')

#Grafico superior derecho
ax[0,1].plot(x,np.cos(x), label = 'cos(x)')
ax[0,1].set_title("Cos(x)")
ax[0,1].set_xlabel('X')
ax[0,1].set_ylabel('Y')
ax[0,1].legend(loc='lower center')

#Grafico inferior izquierdo
ax[1,0].scatter(x,np.random.random(100), label = 'Random Values')
ax[1,0].set_title("Random Values")
ax[1,0].set_xlabel('X')
ax[1,0].set_ylabel('Y')
ax[1,0].legend(loc='center')

#Grafico inferior derecho
ax[1,1].plot(x,np.log(x), label = 'Log(x)')
ax[1,1].set_title("Log(x)")
ax[1,1].set_xlabel('X')
ax[1,1].set_ylabel('Y')
ax[1,1].legend(loc='lower right')


plt.tight_layout()
plt.show()
```

```python
y = np.arange(100)

for x in plt.style.available:
    plt.style.use(x)
    fig,ax = plt.subplots(figsize=(5,5))
    ax.scatter(y,np.random.random(100), label='Random Values')
    ax.plot(y,np.sin(y),label='sin(x)')
    ax.set_title(x,size=20)
    ax.legend()
    plt.tight_layout()
    plt.show()
```
