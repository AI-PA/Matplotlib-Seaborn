## Limpiando Datos

```python
df_races[df_races['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
# Combinando todos los filtros
df_races_filters = df_races[(df_races['Event distance/length'].isin(['50km','50mi'])) & (df_races['Year of event']==2020) & (df_races['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df_races_filters.head(4)
```

```python
df_races_filters.shape
```

```python
# Quitando (USA) del nombre de los eventos
```

```python
df_races_filters['Event name'].str.split('(').str.get(0)
```

```python
df_races_filters['Event name'] =df_races_filters['Event name'].str.split('(').str.get(0)
```

```python
df_races_filters.head(3)
```

```python
# Calculando la edad del atleta
df_races_filters['athlete_age'] = 2020 - df_races_filters['Athlete year of birth']
```

```python
# Quitando H del tiempo del atleta
df_races_filters['Athlete performance'] = df_races_filters['Athlete performance'].str.split(' ').str.get(0)
```

```python
df_races_filters.head()
```

```python
# Eliminando columnas ['Athlete club','Athlete year of birth','Athlete age category']
df_races_filters = df_races_filters.drop(['Athlete club','Athlete year of birth','Athlete age category','Athlete country'], axis=1)
```

```python
df_races_filters.head(5)
```

```python
# Eliminando valores vacíos
```

```python
df_races_filters.isna().sum()
```

```python
df_races_filters[df_races_filters['athlete_age'].isna()==1]
```

```python
df_races_filters=df_races_filters.dropna()
```

```python
df_races_filters.shape
```

```python
# Comprobar si existen duplicados 
df_races_filters[df_races_filters.duplicated()==True]
```

```python
# Reiniciar Indice
```

```python
df_races_filters.reset_index(drop=True)
```

```python
# Corrección de Tipos de Datos
df_races_filters.dtypes
```

```python
df_races_filters['athlete_age'] =df_races_filters['athlete_age'].astype(int)
df_races_filters['Athlete average speed'] =df_races_filters['Athlete average speed'].astype(float)
```

```python
df_races_filters.dtypes
```

```python
# Renombrando columnas 
df_races_filters = df_races_filters.rename(columns={
        'Year of event':'year',
        'Event dates':'race_day',
        'Event name':'race_name',
        'Event distance/length':'race_length',
        'Event number of finishers':'race_number_of_finishers',
        'Athlete performance':'athlete_performance',
        'Athlete gender':'athlete_gender',
        'Athlete average speed':'athlete_average_speed',
        'Athlete ID':'athlete_id'               
})
```

```python
df_races_filters.head()
```

```python
#Reorganizar las columnas
df_races_filters = df_races_filters[['year','race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```

```python
df_races_filters
```

## Graficos

```python
# Graficas
sns.displot(df_races_filters[df_races_filters['race_length']=='50mi']['athlete_average_speed'])
plt.show()
```

```python
sns.violinplot(data=df_races_filters, x='race_length', y="athlete_average_speed",hue='athlete_gender', split=True, inner='quart')
plt.show()
```

```python
sns.lmplot(data=df_races_filters,x='athlete_age',y='athlete_average_speed', hue='athlete_gender')
```

```python
sns.jointplot(data=df_races_filters,x='athlete_average_speed', y='athlete_age',hue='athlete_gender', kind='hist')
plt.show()
```

```python
sns.pairplot(data=df_races_filters, hue='athlete_gender',corner=True,palette='coolwarm')
plt.show()
```
